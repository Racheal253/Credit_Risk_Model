```python
from pyspark.sql import *
from pyspark.sql.functions import* 
from pyspark.sql.types import *
```

```python
retailers_csv = spark.read.csv("Files/synthetic_data/retailers.csv",header=True,inferSchema= True) 
```

```python
display(retailers_csv)
```

```python
retailers_bronze = retailers_csv \
    .withColumn("ingestion_timestamp", current_timestamp()) \
    .withColumn("source_file", lit("retailers.csv"))
```

```python
display(retailers_bronze)
```

```python
retailers_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze_retailers")
```

```python
retailers_nulls = spark.sql("""
    SELECT 
        COUNT(*) - COUNT(retailer_id) as null_retailer_ids,
        COUNT(*) - COUNT(credit_segment) as null_credit_segments
    FROM bronze_retailers
""").collect()[0]

print(f"  Retailers - Null retailer_ids: {retailers_nulls[0]}")
print(f"  Retailers - Null credit_segments: {retailers_nulls[1]}")
```
